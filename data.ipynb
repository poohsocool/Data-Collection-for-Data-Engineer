Create connection to Data Base

```python
import pymysql
connection = pymysql.connect(host=Config.MYSQL_HOST,
                             port=Config.MYSQL_PORT,
                             user=Config.MYSQL_USER,
                             password=Config.MYSQL_PASSWORD,
                             db=Config.MYSQL_DB,
                             charset=Config.MYSQL_CHARSET,
                             cursorclass=pymysql.cursors.DictCursor)
```

After we got data coversion_rate for API and make it to dictionary now we have to set it into Pandas again for easy way to do in the table

```python
conversion_rate = pd.DataFrame(result_conversion_rate)
```

```python
conversion_rate
```

change index to be column date for comfortable to joint with table transaction

```python
conversion_rate = conversion_rate.reset_index().rename(columns={"index": "date"}) #in first column is index but we dont want index so we will make it to column to easily for joining table
conversion_rate[:3]
```

Let's check table both and see how can we use which column to join in this case from conversion_rate table we have date column but in transaction table we have the save column date but
if we watch carefully we will see time-stamp instead therefore we gonna change that column as the same type.

```python
transaction['date'] = transaction['timestamp'] #creating new column but having values as timestamp column
transaction
```

```python
#change date column both table to date types as the same
transaction['date'] = pd.to_datetime(transaction['date']).dt.date
conversion_rate['date'] = pd.to_datetime(conversion_rate['date']).dt.date
transaction.head() # check date colume
```

```python
conversion_rate.head() #check date colume
```

denormalize is a large table that alread joined from every table that we want to know and it's appropriate for analysis.
normalize is suitable for data base it mean that there are many small table. 

```python
#join two table with date column
#in this case, we already joined three table audible_table(from DB), auduble_transaction(from DB) and coversion_rate(from API)
final_df = transaction.merge(conversion_rate, how = "left", left_on = "date", right_on = "date")
final_df.head()
```

Right now we got all data in one table but we have to know price in that time by multpling two column which are Price and conversion_rate, before hand we must chage type Price column to string or change $ it out.

```python
final_df['Price'] = final_df.apply(lambda x: x["Price"].replace("$",""), axis=1) #lambda x is define x to be variable table after : x is mean table and ["Price"] is column / axix =  mean do only that column
final_df['Price'] = final_df['Price'].astype(float) # change type to float
```

```python
final_df.head()
```

```python
final_df.dtypes #to check type all column in the table
```

```python
final_df['THBPrice'] = final_df['Price'] * final_df['conversion_rate']
final_df.head()
```

Another way using def function

```python
def conversion_rate(price, rate):
    return price * rate

#final_df["THBPrice"] = final_df.apply(lambda x: x["Price"] * x["conversion_rate"], axis=1)
#final_df["THBPrice"]

#or use def function
final_df["THBPrice"] = final_df.apply(lambda x: conversion_rate(x["Price"], x["conversion_rate"]), axis=1)
final_df
```

we can drop column that's not neccessary such as date and timstamp
so axis = 1 mean drop column(if axis =0 it can use drop row)

```python
final_df = final_df.drop("date", axis = 1)
```

```python
final_df
```

```python
final_df.to_csv("output.csv", index=False) #we will use index = False to cut index in pandas out because when we export to csv it will show index but we don't need it
```
